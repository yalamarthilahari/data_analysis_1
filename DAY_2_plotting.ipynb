```python
runs=np.array([100,50,91,78,89,25,34,19,9,10])
wickets=np.array([1,0,2,0,3,7,8,9,7,5])
plt.scatter(runs,wickets,color = "g",marker="D") #p-plus,s-square,D-diamond,*-star,<,>.
plt.show()
```

show the scores of two students of five subjects

subjects as xaxis=maths,telugu,hindi,social,science

```python
s1=[56,78,92,95,78]
s2=[77,89,91,23,45]
sub=["maths","science","telugu","hindi","social"]
plt.plot(sub,s1,label="s1",color="g")
plt.plot(sub,s2,label="s2",color="b")
plt.legend()
plt.show()
```

```python
stu1=[56,78,92,95,78]
stu2=[77,89,91,23,45]
sub=["maths","science","english","telugu","social"]
plt.subplot(2,1,1)
plt.plot(sub,stu1,label="student1",color="b")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,stu2,label="student2",color="k")
plt.legend()
plt.show()
```

```python
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("Revenues of Company A",A)
print("Revenues of Company B",B)
profA=np.diff(A)
print("profits of A",profA)
profB=np.diff(B)
print("profits of B",profB)
Y=["19-20","20-21","21-22","22-23"]
plt.subplot(1,2,1)
plt.bar(Y,profA,color='k',label="profit A")
plt.title("profits of  A")
plt.legend(loc="upper right")
plt.subplot(1,2,2)
plt.bar(Y,profB,color="g",label="profit B")
plt.legend(loc="best")
plt.title("profits of B")


```

```python
a=np.array([25,60,5,10])
label=["aiml","python","pandas","numpy"]
plt.pie(a,labels=label)
plt.show()
```

```python
#explode
a=np.array([25,60,5,10])
label=["aiml","python","pandas","numpy"]
explo=[0.1,0,0,0.1]
plt.pie(a,labels=label,explode=explo,startangle=180,shadow=True)
plt.legend()
plt.show()
```
