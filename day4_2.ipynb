```python
import pandas as pd
```

```python
a=('https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1')
req=requests.get(a)
```

```python
b=BeautifulSoup(req.text,"html")
print(b)
```

```python
#tr-table row
#th-table head #td-tabledata
#tb-tablebody
#table parent
table=b.find('table')
print(table)
```
